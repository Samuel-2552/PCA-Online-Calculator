```python
import pandas as pd
import numpy as np
```

```python
path="iris_dataset.csv"
#to get the number of colums
columns=int(input()) #get the number of columns and name of the columns from the user
```

```python
#to get the column names from user
lst=[]
for i in range(0,columns):
    column_names=input()
    lst.append(column_names)
lst
```

```python
df=pd.read_csv(path)
df.head(10)
```

```python
missing_values=df.isnull().sum()
missing_values
#ms=df.isnull()
#ms
```

```python
lst_input=lst[:-1]
lst_input
```

```python
#filling missing values with mean
for cn in lst_input:
    mean_value=df[cn].mean()
    df[cn].fillna(value=mean_value, inplace=True)
df.head(10)
```

```python
#for standerdization
from sklearn.preprocessing import StandardScaler
scaled_values=StandardScaler().fit_transform(df[lst_input])
scaled=pd.DataFrame(scaled_values,columns=lst_input)
scaled
```

```python
#for normalization
from sklearn import preprocessing
scaled_values=preprocessing.normalize(df[lst_input])
scaled=pd.DataFrame(scaled_values,columns=lst_input)
scaled
```

```python
#for removing outliers
def replace_outlier(val, mean, std):
    if val > mean + 3*std:
        return mean + 3*std 
    elif val < mean - 3*std:
        return mean - 3*std
    return val

for col in lst_input:
    mean = df[col].mean()
    std_dev = df[col].std(axis=0)
    df[col] = df[col].map(lambda x: replace_outlier(x, mean, std_dev))

df.head(10)
```
